```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
#read canda per capita income data
df = pd.read_csv("canada_per_capita_income.csv")
df
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('Income')
plt.scatter(df.year, df.Income, color="red", marker="+")
```

```python
new_df = df.drop('Income', axis='columns')
new_df
```

```python
income = df.Income
income
```

```python
#Create Regresssion Model
reg = linear_model.LinearRegression()
reg.fit(new_df, income)
```

```python
reg.predict([[2020]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
year_df = pd.read_csv("year.csv")
year_df
```

```python
predict_income = reg.predict(year_df)
predict_income
```

```python
year_df['Income'] = predict_income
year_df
```

```python
year_df.to_csv("incomeprediction.csv")
```
